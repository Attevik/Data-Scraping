```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from datetime import date
import numpy as np
from datetime import date
 
```

```python
url='https://akcern.am/en/category?city=&property-type=72&realty-type=&page=1&per-page=15'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text, 'lxml')
soup
```

```python
#finds all links
List_Links=[]
link_for_all=soup.find_all('a', {'class':'property-img'})
for i in link_for_all:
    i.get('href')
    full_link2='https://akcern.am/en'+i.get('href')
    List_Links.append(full_link2)
    
print(List_Links)
len(List_Links)
```

```python
#finding all codes
List_codes=[]
all_codes=soup.find_all('div', {'class':'card-code'})
for i in all_codes:
    a=i.text.strip()
    List_codes.append(a)
print(List_codes)
```

```python
List_sales=[]
For_sale_rent=soup.find_all('h1', {'class': "title"})
for i in For_sale_rent:
    Sales=i.text.strip()
    List_sales.append(Sales)
#print(List_sales)
Final_list=[]
for i in List_sales:
    sliced=i[0:8]
    Final_list.append(sliced)
print(Final_list)
```

```python
List_prices=[]
all_prices=soup.find_all('div', {'class':'price-ratings-box'})
for i in all_prices:
    b=i.text.strip()
    List_prices.append(b)
print(List_prices)
```

```python
Total=[]
Total_area=soup.find_all('ul', {'class':'facilities-list clearfix'})
for i in Total_area:
    d=i.text.strip()
    Total.append(d)
print(Total)
len(Total)
```

```python
Only_area=[]
for i in range(len(Total)):
    k=re.findall('Total\sarea\s\d*\s\w\s\d', Total[i])
    Only_area.append(k)
Without_brackets2= [i[0] for i in Only_area]
print(Without_brackets2)
```

```python
Only_Floor=[]
for i in range(len(Total)):
    t=re.findall('(Floor\s\d*)', Total[i])
    Only_Floor.append(t)
Without_brackets1= [i[0] for i in Only_Floor]
print(Without_brackets1)
```

```python
#'(Count\s\w\w\s\w\w\w\w\w\s\d*)'
Only_rooms=[]
for i in range(len(Total)):
    r=re.findall('(Count\sof\srooms\s\d*)', Total[i])
    Only_rooms.append(r)
Without_brackets= [i[0] for i in Only_rooms]
print(Without_brackets)
```

```python
List_dates=[]
all_dates=soup.find_all('div', {'class':'footer'})
for i in all_dates:
    l=i.text.strip()
    List_dates.append(l)
print(List_dates)
 
```

```python
Location_list=[]
Location=List_Links
for i1 in List_Links:
    page1=requests.get(i1)
    soup1=BeautifulSoup(page1.text, 'lxml')
    locations=soup1.find('div', {'class':'detail-location'})
    Text_loc=locations.text.strip()
    Location_list.append(Text_loc)
print(Location_list)
    
    
```

```python
all_products=soup.find_all('div', {'class':'col-xl-4 col-lg-6 col-md-4 col-sm-6 col-12'})

data=pd.DataFrame({'Link':[], 'Code':[],'For_rent_sale':[], 'Price':[], 'Area':[],'Floor':[], 'Rooms':[], 'Dates':[] })

for i in all_products:
    link_for_all=i.find_all('a', {'class':'property-img' })[0].get("href") 
    full_link2='https://akcern.am/en'+link_for_all
    codes=i.find_all('div', {'class':'card-code'})[0].text.strip()
    For_sale_rent=i.find_all('h1', {'class': "title"})[0].text.strip()[0:8]
    prices=i.find_all('div', {'class':'price-ratings-box'})[0].text.strip()
    area=i.find_all('ul', {'class':'facilities-list clearfix'})[0].text.strip()
    total_area = str(re.findall('Total\sarea\s\d*\s\w\s\d', area)).strip('[]')
    floor = str(re.findall('(Floor\s\d*)', area)).strip('[]')
    room = str(re.findall('(Count\sof\srooms\s\d*)', area)).strip('[]')
    date=i.find_all('div', {'class':'footer'})[0].text.strip()
    data = data.append({'Link':full_link2, 'Code':codes,'For_rent_sale':For_sale_rent, 'Price':prices,
                        'Area':total_area,'Floor':floor, 'Rooms':room, 'Dates':date }, ignore_index= True)
    
data['location']= Location_list
data
```

```python
data=pd.DataFrame({'Link':[], 'Code':[],'For_rent_sale':[], 'Price':[], 'Area':[],'Floor':[], 'Rooms':[], 'Dates':[] })

i = 1
while True:
    url = f'https://akcern.am/en/category?city=&property-type=72&realty-type=&page={i}&per-page=15'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')
    postings = soup.find_all('div', {'class':'col-xl-4 col-lg-6 col-md-4 col-sm-6 col-12'})
    for post in postings:
        link_for_all=post.find_all('a', {'class':'property-img' })[0].get("href") 
        full_link2='https://akcern.am/en'+link_for_all
        codes=post.find_all('div', {'class':'card-code'})[0].text.strip()
        For_sale_rent=post.find_all('h1', {'class': "title"})[0].text.strip()[0:8]
        prices=post.find_all('div', {'class':'price-ratings-box'})[0].text.strip()
        area=post.find_all('ul', {'class':'facilities-list clearfix'})[0].text.strip()
        total_area = str(re.findall('Total\sarea\s\d*\s\w\s\d', area)).strip('[]')
        floor = str(re.findall('(Floor\s\d*)', area)).strip('[]')
        room = str(re.findall('(Count\sof\srooms\s\d*)', area)).strip('[]')
        date=post.find_all('div', {'class':'footer'})[0].text.strip()
             
        data= data.append({'Link':full_link2, 'Code':codes,'For_rent_sale':For_sale_rent, 'Price':prices,
                        'Area':total_area,'Floor':floor, 'Rooms':room, 'Dates':date }, ignore_index= True)
    print(url)
    i = i+1
    if url == 'https://akcern.am/en/category?city=&property-type=72&realty-type=&page=320&per-page=15':
        break
        
```
