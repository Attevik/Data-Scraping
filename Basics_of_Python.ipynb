```python
#Question2
def summary (x):
    Min=str(min(x))
    Max=str(max(x))
    SD=round(np.std(x), 2)
    Mean= round(np.mean(x), 2)
    CV=round(np.std(x)/np.mean(x), 2)
    Length=str(len(x))
    data_for_func = pd.DataFrame({'Metric':['Min','Max','SD','Mean','CV','Length'], 'Value':[Min, Max, SD, Mean, CV, Length]})
    
    return data_for_func


                   
```
